```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("Chapman.dat", dtype=str)
Header = (data[0,:]).tolist()
solTchem = (data[1:,:]).astype(float)
```

```python
print(Header)
```

```python
for var in Header:
    var_indx=Header.index(var)
    plt.figure()
    plt.plot(solTchem[:,1]/3600,solTchem[:,var_indx] )
    plt.xlabel('Time [hr]')
    plt.ylabel(var)
```
