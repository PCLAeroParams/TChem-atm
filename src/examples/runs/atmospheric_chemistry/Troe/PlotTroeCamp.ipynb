```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("Troe.dat", dtype=str)
Header = (data[0,:]).tolist()
solTchem = (data[1:,:]).astype(np.float)
```

```python
campsol = np.loadtxt("troe_results.txt")
```

```python
A_indx = Header.index('A')
B_indx = Header.index('B')
C_indx = Header.index('C')
t_indx = Header.index('t')
```

```python
print(np.shape(data))
```

```python
plt.figure()
plt.plot(solTchem[:,t_indx],solTchem[:,A_indx], label="A")
plt.plot(solTchem[:,t_indx],solTchem[:,B_indx], label="B")
plt.plot(solTchem[:,t_indx],solTchem[:,C_indx], label="C")
plt.xlabel('Time [s]')
plt.ylabel('Concentration [mol/m3]')
plt.legend(loc='best')
```

```python
plt.figure()
plt.plot(campsol[:,0],campsol[:,1],label='CAMP')
n=1000
plt.plot(solTchem[:n,t_indx],solTchem[:n,A_indx],"*",label='TChem')
plt.xlabel('Time [s]')
plt.ylabel('Concentration of A [mol/m3]')
plt.legend(loc='best')
```

```python
plt.figure()
plt.plot(campsol[:,0],campsol[:,3],label='CAMP')
plt.plot(solTchem[:n,t_indx],solTchem[:n,B_indx],"*",label='TChem')
plt.xlabel('Time [s]')
plt.ylabel('Concentration of B [mol/m3]')
plt.legend(loc='best')
```

```python
plt.figure()
plt.plot(campsol[:,0],campsol[:,5],label='CAMP')
plt.plot(solTchem[:n,t_indx],solTchem[:n,C_indx],"*",label='TChem')
plt.xlabel('Time [s]')
plt.ylabel('Concentration of B [mol/m3]')
plt.legend(loc='best')
```
